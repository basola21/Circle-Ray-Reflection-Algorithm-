```python
import math
```

```python
# first we are given the entry point (1 , 9.95)
# and the refection point on the circle is given (8,-6)
# using the two points find the slope of the first line = m

x1 = 1
y1 = 9.95
x2 = 8
y2 = -6

m = (y2-y1)/(x2-x1)

print (m)
```

```python
# finding the slope of the tangent at the intersection mT
# since we know from the question that any tangent to the circle is = -x/y

mT = -x2/y2

print(mT)
```

```python
# getting the normal to the tangent 
# so that we can then get the angle between the normal to the tangent and the orignal line 
# this angle is equal to the angle between the reflected line and the normal on the tangent 
# mN is the slope of the normal on the tangent
# mT * mN = -1 --> mN = -1/mT

mN = -1/mT

print(mN)
```

```python
# getting the angle between two slopes using the tan rule 

# documentation on the tan rule 

tan0 = (m-mN)/(1+m*mN)

print(tan0)
```

```python
# then using the same rule but now having the Tan0 to get the slope of the reflected light 
# mR is the slope of the reflected line

mR = (mN-tan0)/(1-tan0*mN)

print(mR)
```

```python
# solving the intersection between the line and the circle 
# will give a quadratic equation 
# slove the quadratic equation to get the new intersection 

a = 1 + (mR**2)
b = (-2*mR**2*x2+2*y2*mR)
c = -(2*y2*mR*x2)+(y2**2) + (mR**2 * x2**2)-100 

print(a,b,c)

quad1 = (-b-math.sqrt(b-4*a*c))/(2*a)
quad2 = (-b+math.sqrt(b-4*a*c))/(2*a)

print(quad1,quad2)
```

```python
#getting the new y using quad 1 and quad 2 


if  math.isclose(x2,quad1):
#then get the y for quad 2 and start the reflection all again
    x1 = x2
    y1 = y2
    x2 = quad2
    y2 = mR*(x2-x1)+y1

    print(x1,y1,x2,y2)
                
else:
#get the y of quad 1 and start the reflection problem 
    x1 = x2
    y1 = y2
    x2 = quad1
    y2 = mR*(x2-x1)+y1
    print(x1,y1,x2,y2)
```
